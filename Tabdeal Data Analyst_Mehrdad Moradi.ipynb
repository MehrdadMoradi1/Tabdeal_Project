```python
#necessary modules
import numpy as np
import pandas as pd
import dateutil
from datetime import datetime
from pytz import utc
```

```python
data=pd.read_csv('data tabdeal.csv')
```

```python
data.head()
```

```python
#do we have missing values?
data.isna().sum()
```

```python
#converting date to day, month, year
data['day'] = pd.to_datetime(data['Date']).dt.strftime('%d')
data['month'] = pd.to_datetime(data['Date']).dt.strftime('%m')
data['year'] = pd.to_datetime(data['Date']).dt.strftime('%y')
```

```python
data = data.astype({"day": int, "month": int, "year": int}, errors='raise') 
```

```python
#for every row, we calculate the total revenue  
data['total revenue']=data['Usdt Commission']+(data['Irt Commission']/data['Usdt Irt Price'])-(data['Rebate']+data['Referrer Rebate'])
```

```python
total_revenue=data['total revenue'].sum(axis=0)
unique_customers=data['user id'].nunique()
print('revenue per cutomer for the given period is %f Usdt' %(total_revenue/unique_customers))
print('revenue per cutomer per month is %f Usdt' %(total_revenue/(6*unique_customers)))
```

```python
#date to date in days
len_month=[31,28,31,30,31,30,31,31,30,31,30,31]

def total(x):
    s=0
    for i in range(x-1):
        s+=len_month[i]
    return s
#data['length of month']=data['month'].apply(lambda x: 31 if x in [5,7,8] else 30)
#data['time']=100*data['month']+data['day']


data['date in days']=data['month'].apply(lambda x: total(x))
data['date in days']=data['date in days']+data['day']
```

```python
data.head()
```

```python
#creating a dataframe for churn rate calcualtion
churn=pd.DataFrame([i for i in range(4,10,1)],columns=['month'])
churn
```

```python
#churn rate calculation
for i in range(4,10,1):
    
    s1=data[(data['date in days']<=total(i)) & (data['date in days']>total(i)-30)]['user id']
    s2=data[(data['date in days']<=total(i+1)) & (data['date in days']>total(i+1)-30)]['user id']
    s3=pd.Series(list(set(s1) & set(s2)),dtype='float64')
    
    churn.loc[i-4,'churned']=s1.nunique()-s3.nunique()
    churn.loc[i-4,'active + new + reactive']=s2.nunique()
    churn.loc[i-4,'churn rate']=str(round(100*(((s1.nunique()-s3.nunique()))/(s2.nunique())),1))+'%'
    churn.loc[i-4,'100% - churn rate']=str(round(100*(1-(((s1.nunique()-s3.nunique()))/(s2.nunique()))),1))+'%'
```

```python
#set 'month' as index
churn=churn.set_index('month')
```

```python
#showing churn in a proper manner
churn.style.background_gradient(cmap ='viridis')\
        .set_properties(**{'font-size': '20px'})
```

```python
#CLV= ACL (average customer life span in month)* APF (average purchase frequency per month) * APV (average purchase value)
APV = total_revenue/(data.shape[0]-data[data['total revenue']==0].shape[0])
APF = (data.shape[0])/(6*unique_customers)
ACL = ((data.groupby(['user id'])['date in days'].max()-data.groupby(['user id'])['date in days'].min()).mean()/30)

CLV = APF * APV * ACL

print('CLV is %f Usdt' %CLV)
```

```python
#creating a dataframe for cohort
lst=[i for i in range(1,32,1)]
cohort = pd.DataFrame(lst, columns =['day of month 5'])
```

```python
#creating a dataframe for the statistics of cohort 
cohort_stat=pd.DataFrame(columns=['which day?'],index=['mean','median','mode','standard deviation'])
cohort_stat
```

```python
cohort
```

```python
#completing the last two dataframes
for i in range(0,32,1):
    j=31-i
    
    lst1=[]
    for s in range(j):
        d=s+i
        cohort.loc[s,'day %i' %i]=str(round(100*(pd.Series(list(set(data[(data['month']==5) & (data['day']==s+1)]['user id']) & 
        set(data[(data['month']==5) & (data['day']==d+1)]['user id'])),dtype='float64').nunique())/(data[(data['month']==5) & (data['day']==s+1)]['user id']).nunique(),0))+'%'
        
        lst1.append(round(100*(pd.Series(list(set(data[(data['month']==5) & (data['day']==s+1)]['user id']) & 
        set(data[(data['month']==5) & (data['day']==d+1)]['user id'])),dtype='float64').nunique())/(data[(data['month']==5) & (data['day']==s+1)]['user id']).nunique(),0))
        cohort_stat.loc['mean','day %i' %i]=np.mean(lst1)
        cohort_stat.loc['median','day %i' %i]=np.median(lst1)
        cohort_stat.loc['mode','day %i' %i]=(max(set(lst1), key = lst1.count))
        cohort_stat.loc['standard deviation','day %i' %i]=(sum([((x - (sum(lst1) / len(lst1))) ** 2) for x in lst1]) / len(lst1)) ** 0.5
```

```python
#replace NAN with ""
cohort=cohort. replace(np. nan,'',regex=True)
#setting index
cohort=cohort. set_index("day of month 5")
```

```python
#showing cohort in a proper manner
cohort.style.background_gradient(cmap ='viridis')\
        .set_properties(**{'font-size': '20px'})
```
